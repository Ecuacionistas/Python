```python
x = np.arange(-100,100)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('Cantidad de horas de uso')
plt.ylabel('Costo $/hr')
plt.title('Consumo de dinero por hora trabajada de un motor')
plt.show()
#forma de plotear una funcion y de agregarle titulos a los ejes y al grafico
```

```python
fig = plt.figure(figsize=(12,8) , dpi =100)
#creo una figura y la guardo en fig
axis = fig.add_axes([0,0,1,1])
#le indico a la figura el eje desde donde empieza y desde donde va
axis2 = fig.add_axes([0.5,0.5,0.4,0.4])
axis.set_xlabel('Cantidad de hr')
axis.set_ylabel('consumo $')
axis.set_title('consumo de motor')
#le agrega los ejes y los personalizo
axis.set_xlim(0,100)
axis.set_ylim(0,2100)
#coloco los limites a los ejes
axis.plot(x,y)
#le agrega los ejes y los personalizo
axis2.set_title('curva caracteristica en rpm')
axis2.plot(x,x**3)
plt.show
```
